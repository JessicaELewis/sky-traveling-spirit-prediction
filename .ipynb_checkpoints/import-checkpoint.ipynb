# Sky Traveling Spirit Prediction

```python
import pandas as pd
```

```python
spirits = pd.read_csv('data/sky_traveling_spirits.csv')
spirits.head()
```

```python
spirits.rename(columns={'Visit#':'Visit', 'Last Visit':'Visit Number'}, inplace=True)
```

```python
spirits.head()
```

```python
spirits['Visit'].replace('[a-z]','',regex=True, inplace=True)
```

```python
spirits.Visit.unique()
```

```python
spirits = spirits[spirits.Visit != 'E*']
```

```python
spirits.Visit = spirits.Visit.astype('int64')
```

```python
spirits.set_index('Visit').sort_index()
```

```python
spirits.Date = pd.to_datetime(spirits.Date)
```

```python
spirits.head()
```

```python
spirits['Previous Visit'] = spirits['Visit Number'].str.extract(r'\(([^\)]+)\)', expand=True)
spirits.head()
```

```python
spirits['Visit Number'] = spirits['Visit Number'].str.extract(pat='(^[1-9]+)')
```
